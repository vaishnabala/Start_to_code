```python
some_sent = "There is a space at the end of this sentence.    "
print("\n Prints the sentence")
print(some_sent)

print("\nuses rstrip")
print(some_sent.rstrip())  #removes trailing spaces

print("\n Prints the sentence")
some_sent = "               There is a space at the end of this sentence."
print(some_sent)

print("\nuses lstrip")
print(some_sent.lstrip())  #removes leading spaces

some_sent = "           There is a space at the end of this sentence.    "
print("Prints the sentence")
print(some_sent)

print("\nuses strip")
print(some_sent.strip())   #removes leading and trailing spaces

print("\nUses Replace function")
print(some_sent.replace("space", "gap"))  #replaces a word with another word

print("\nUses Upper case and Lower case functions")
print(some_sent.upper())   #converts to uppercase
print(some_sent.lower())   #converts to lowercase

print("\nsentence, rstrip, lstrip and strip with specified characters")
some_sent = "%%%%There is a space at the end of this sentence%%%%"
print(some_sent)
print(some_sent.rstrip("%"))  #removes specified characters from both ends
print(some_sent.lstrip("%"))  #removes specified characters from both ends
print(some_sent.strip("%"))   #removes specified characters from both ends


print("\nCan use bothe rstrip and lstrip together")
some_sent = "%%%%There is a example*********"
print(some_sent)
print(some_sent.rstrip("*").lstrip("%")  )#removes specified characters from both ends)    
```
